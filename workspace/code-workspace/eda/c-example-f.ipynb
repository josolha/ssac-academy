```python
import numpy as np
import pandas as pd
import requests
import time
from bs4 import BeautifulSoup
```

```python
page_no = 1

url = "https://news.daum.net/breakingnews/digital?page={0}"

resp = requests.get(url.format(page_no))
```

```python
print( resp.status_code )
resp.text[:300]
```

```python
soup = BeautifulSoup(resp.text, "html.parser")
```

```python
lis = soup.select('#mArticle > .box_etc li')
# print( len(lis) )
# lis[:3]
```

```python
rows = []
for li in lis:
    link = li.select_one('.cont_thumb strong a')
    desc = li.select_one('.cont_thumb .desc_thumb span')
    # print("[", link.get_text().strip(), desc.get_text().strip(), link['href'], "]")
    rows.append([link.get_text().strip(), desc.get_text().strip(), link['href']])
```

```python
page_no = 1
reg_date = '20210201'
url = "https://news.daum.net/breakingnews/digital?page={0}&regDate={1}"
rows = []

while True:
    resp = requests.get(url.format(page_no, reg_date))

    soup = BeautifulSoup(resp.text, "html.parser")

    lis = soup.select('#mArticle > .box_etc li')

    if lis == None or len(lis) == 0:
        print("End of List")
        break

    for li in lis:
        link = li.select_one('.cont_thumb strong a')
        desc = li.select_one('.cont_thumb .desc_thumb span')
        # print("[", link.get_text().strip(), desc.get_text().strip(), link['href'], "]")
        rows.append([link.get_text().strip(), desc.get_text().strip(), link['href']])

    page_no += 1

    time.sleep(1)
```
